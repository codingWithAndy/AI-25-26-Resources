## Task 3 – Arithmetic operators

```python
a = 15
b = 4

print("addition:", a + b)
print("subtraction:", a - b)
print("multiplication:", a * b)
print("division:", a / b)
print("integer division:", a // b)
print("modulus:", a % b)
print("exponentiation:", a ** b)
```

## Task 5 – Lists

```python
ai_topics = ["search", "machine learning", "computer vision", "NLP"]

print(ai_topics)
print(ai_topics[0])
print(ai_topics[-1])

ai_topics.append("robotics")
print(ai_topics)
```

## Task 6 – List indexing and slicing

```python
topics = ["python", "data", "logic", "models", "agents"]

print(topics[1])
print(topics[2:4])
print(topics[::-1])
```

## Task 7 – Dictionaries

```python
student = {
    "name": "Andy",
    "level": 5,
    "module": "Artificial Intelligence",
    "skills": ["problem solving", "communication", "coding"]
}

print(student)
print(student["module"])
print(student["skills"][0])
```

## Task 8 – Selection with `if`, `elif`, `else`

```python
mark = 68

if mark >= 70:
    print("First")
elif mark >= 60:
    print("Upper Second")
elif mark >= 50:
    print("Lower Second")
else:
    print("Needs improvement")

mark = 45

if mark >= 70:
    print("First")
elif mark >= 60:
    print("Upper Second")
elif mark >= 50:
    print("Lower Second")
else:
    print("Needs improvement")
```

## Task 9 – `for` loops

```python
for i in range(1, 11):
    print(i)

for item in ["AI", "Python", "Data", "Models"]:
    print(item)
```

## Task 10 – `while` loops

```python
i = 5

while i >= 1:
    print(i)
    i -= 1

print("Done")
```

## Task 11 – Functions

```python
def greet_student(name):
    print(f"Welcome to AI, {name}")

greet_student("Andy")
```

## Task 12 – Functions with return values

```python
def square_number(number):
    return number ** 2

result = square_number(6)
print(result)
```

## Task 13 – Mini problem: total and average

```python
marks = [62, 75, 68, 71, 59]

print("total:", sum(marks))
print("average:", sum(marks) / len(marks))
print("highest:", max(marks))
print("lowest:", min(marks))
```

## Task 14 – List comprehension

```python
squares = [i ** 2 for i in range(10)]
print(squares)
```

## Task 15 – Challenge task

```python
name = "Andy"
hours = 8

print(f"{name} studied for {hours} hours this week.")

if hours >= 10:
    print("Excellent effort")
elif hours >= 5:
    print("Good start")
else:
    print("Try to build a regular study habit")
```
