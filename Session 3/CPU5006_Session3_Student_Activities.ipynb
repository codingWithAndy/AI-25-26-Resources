## Slides: First Python Rule-Based Example  
### Activity 5: Predict the output

Read the code and predict the outputs before running it.


```
class RuleBasedAI:
    def respond(self, message):
        message = message.lower()
        if "hello" in message:
            return "Hello! How can I assist you today?"
        elif "how are you" in message:
            return "I'm an AI, I don't have feelings, but I'm here to help you."
        elif "what is your name" in message:
            return "I'm a Rule Based AI."
        else:
            return "I'm sorry, I didn't understand that."

ai = RuleBasedAI()

print(ai.respond("Hello there"))
print(ai.respond("What is your name?"))
print(ai.respond("Tell me a joke"))
```
