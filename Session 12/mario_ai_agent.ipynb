## 3. Train the Mario Agent

```python
model = DQN(
    'CnnPolicy',
    env,
    learning_rate=1e-4,
    buffer_size=50000,
    learning_starts=1000,
    batch_size=32,
    gamma=0.99,
    train_freq=4,
    target_update_interval=1000,
    verbose=1
)

model.learn(total_timesteps=100000)
model.save("mario_dqn")
print("Training complete! Model saved as mario_dqn.")
```

## 4. Watch the Agent Play

```python
import cv2

model = DQN.load("mario_dqn", env=env)

state = env.reset()
done = False

while not done:
    action, _ = model.predict(state)
    state, reward, done, info = env.step(action)
    env.render()

env.close()
print("Agent finished running.")
```
