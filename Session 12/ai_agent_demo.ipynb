# 🧠 Simple AI Agent Demo – Grid World
This notebook demonstrates a simple AI agent using the **Sense → Think → Act** loop.

## 1. Define the Agent Class

```python
import time

class GridWorldAgent:
    def __init__(self, start, goal):
        self.position = start
        self.goal = goal

    def sense(self):
        px, py = self.position
        gx, gy = self.goal
        return {
            'goal_left': gx < px,
            'goal_right': gx > px,
            'goal_up': gy < py,
            'goal_down': gy > py
        }

    def think(self, perception):
        if perception['goal_right']:
            return 'MOVE_RIGHT'
        if perception['goal_left']:
            return 'MOVE_LEFT'
        if perception['goal_up']:
            return 'MOVE_UP'
        if perception['goal_down']:
            return 'MOVE_DOWN'
        return 'STAY'

    def act(self, action):
        x, y = self.position
        if action == 'MOVE_RIGHT':
            self.position = (x + 1, y)
        elif action == 'MOVE_LEFT':
            self.position = (x - 1, y)
        elif action == 'MOVE_UP':
            self.position = (x, y - 1)
        elif action == 'MOVE_DOWN':
            self.position = (x, y + 1)

    def step(self):
        perception = self.sense()
        action = self.think(perception)
        self.act(action)
        return action
```

## 2. Display Helper Function

```python
def display(agent):
    grid_size = 10

    for y in range(grid_size):
        row = ''
        for x in range(grid_size):
            if (x, y) == agent.position:
                row += ' A '
            elif (x, y) == agent.goal:
                row += ' G '
            else:
                row += ' . '
        print(row)

    print('\nAgent position:', agent.position)
```

## 3. Run the Agent

```python
agent = GridWorldAgent(start=(0, 0), goal=(7, 5))

while agent.position != agent.goal:
    display(agent)
    action = agent.step()
    print('Action:', action)
    time.sleep(4)
    print('\n')

display(agent)
print('Goal reached!')
```
